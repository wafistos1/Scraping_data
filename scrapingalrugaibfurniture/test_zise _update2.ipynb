```python
list_car = [
       'Dimension:','Top', 'Dimensions', '(Confirmed):Inches:',  'Opening', 'drawer', 'drawer', '(lined)', 'interior', '"','Inches', 'Inches', 'inches', 'inche', 'Inche','in','-', 'd-', 'h-', 'w-', '', 'بوصة', 'ارتفاع', 'عمق', 'الأبعاد','عرض','W', 'w', ':', '"', 'x', 'X', '"',
        'D', 'H', 'h', 'd', 'بوصة:', 'بوصة', 'ة' , 'ﺔﺻﻮﺑ', ' ﻖﻤﻋ', 'في', '(2)'
    ]
def delete_all_carecter(text):
    print(text.split(' '))
    
    for car in text.split(' '):
        
        if car in list_car:
            print(car)
            text = text.replace(car, '')
    return text.replace('"', '').replace('Dimensions:', '').replace(' (بوصة)', '').replace('raer', '').replace('الابعاد:', '').replace('W', '').replace('D', '').replace('H', '').replace('h', '').replace('Inces:', '').replace('w', '').replace('-', '').replace('d', '').replace('h', '').strip()
```
