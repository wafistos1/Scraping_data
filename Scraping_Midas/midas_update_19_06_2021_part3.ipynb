```python
count = 0
list_images = []
list_input_title = []
list_input_price = []
list_input_images1 = []
list_input_images2 = []
list_input_images3 = []
list_input_images4 = []
list_input_price_was = []
list_input_model = []
list_input_size = []
list_input_color = []
list_input_sku = []
list_input_style = []
list_input_in_stock = []
list_input_mater = []
list_input_link_url = []
list_input_category1 = []
list_input_category2 = []
list_input_category3 = []
list_input_width = []
list_input_height = []
list_input_length = []
list_input_note = []
list_input_overview = []
list_input_product_size = []
list_concatenation = []
list_visibility = []
list_tax_class_name = []
note = ''

#ForLoop main
for product in list_scrape:
    time.sleep(1)
    try:
        driver.get(product)
        time.sleep(3)
        
    except:
        
        continue
    try:
        product_name = driver.find_element_by_class_name('product-name').text.split('(')[0].replace('\n', '')
        try:
            if 'هذا المنتج سيكون' in driver.find_element_by_class_name('product-name').text.split('(')[1]:
                note = driver.find_element_by_class_name('product-name').text.split('(')[1].replace(')', '')
            else:
                note = ''
        except:
            note = ''
        
        link_url = str(product)

        in_stock1 = driver.find_element_by_class_name('availability').text
        if in_stock1 == 'متوفر':
            in_stock  = 1
        else:
            in_stock = 0

        sku = 'DAS-'+driver.find_element_by_class_name('sku').text[4:]

        old_price = driver.find_elements_by_class_name('price')[0].get_attribute('innerHTML').replace('SAR', '').replace('\n', '').replace(' ', '')

        price = driver.find_elements_by_class_name('price')[1].get_attribute('innerHTML').replace('SAR', '').replace('\n', '').replace(' ', '')

        combination = driver.find_element_by_class_name('std').text
    except:
        continue
    combination_product_size_list = combination.split('\n')
    combination_product_size_list1 = driver.find_element_by_class_name('std').text.split('\n\n')
    try:
        combination = combination_product_size_list[1].replace('♦', '')
    except:
        combination = combination_product_size_list[0].replace('♦', '')
    #product_size = combination_product_size_list[4].replace('♦', '')
    
 # CATEGORY  *******************************************   
 
    try:
        category1 =  ''#driver.find_element_by_class_name('category47').text
    except:
        category1 = ''
    try:
        category2 =  ''#driver.find_element_by_class_name('category52').text
    except:
        category2 = ''
    try:
        category3 = ''#driver.find_element_by_class_name('').text
    except:
        category3 = ''
    print('Cat1:', category1)
    print('Cat2:', category2)
    print('Cat3:', category3)

 # MATTER  *******************************************   
  
    mater = return_details('مواد التصنيع', combination_product_size_list1).replace("\n", " ").replace(':', '').replace('♦', '')
    if mater == '':
        mater = return_details('المواد الرئيسية', combination_product_size_list1).replace("\n", " ").replace(':', '').replace('♦', '')
    if mater == '':
        for  matter in combination_product_size_list1:
            if matter.find('Materials') == 0:
                
                mater = matter.replace('Materials', '').replace(':', '').replace('♦', '')
    if mater == '':
        mater = '__EMPTY__VALUE__'
    print('Mater', mater)
# DETAIL  *******************************************   
    
    detail = return_details('تفاصيل المنتج', combination_product_size_list1).replace("\n", " ").replace(':', '').replace('♦', '')
    if detail == '':
        for dt in combination_product_size_list1:


            if dt.find('Combination') == 0:
                print('coucou')
                detail = dt.replace('♦ ', '').replace('Combination', '').replace(':', '').replace('\n', '')
    print('Detail: ',detail)   
# COLOR  *******************************************   

    color = return_details('اللون', combination_product_size_list1).replace("\n", " ").replace('♦', '').replace(':', '')
    if color == '':
        # Additional
        try:
            additional_tabbed = driver.find_element_by_id('product_tabs_additional_tabbed')
            additional_tabbed.click()
            time.sleep(1)
            color = driver.find_element_by_id('product-attribute-specs-table').text.replace('اللون', '')
            
        except:
            color = ''
    if color == '':
        color = '__EMPTY__VALUE__'
    print('Color: ', color)
# SIZE  *******************************************   
    size = return_details('حجم المنتج', combination_product_size_list1).replace("\n", " ").replace('♦', '').replace(':', '').replace('مصباح طاولة', '').replace('مصباح أرضي', '').replace('مصباح أرضي:', '').replace('ثريا سقف', '')
    
    
                
    
    if size.find('Table Lamp') == 0:
        size = size.replace('Table Lamp', '')
    if size == '':
        for size_find in combination_product_size_list1:
            if size_find.find('الحجم') == 0:
                size = size_find.replace('الحجم', '').replace(':', '').replace('\n', ''). replace('♦', '').replace('  ', '')
            elif size_find.find('Product Size') == 0:
                size = [toto for toto in combination_product_size_list1 if toto.find('Product Size') == 0][0]
    toto = size.split('*')
    print(toto)
    #Width
    try:
        width1 = [x for x in toto if 'عرض' in x or 'العرض' in x]
        width2 = width1[0].split(' ')
        width3 =  [x for x in width2 if x.isdigit()]
        width =  str(width3[0])
        print('width',width)
    except:
        print('Pas de width')
        width = ''
    #height
    try:
        height1 = [x for x in toto if 'إرتفاع' in x or ' ارتفاع'in x or 'الارتفاع' in x  ]
        height2 = height1[0].split(' ')
        height3 = [x for x in height2 if x.isdigit()]
        height = str(height3[0])
        print('height',height)
    except:
        print('Pas de height')
        height = ''

    #length
    try:
        length1 = [x for x in toto if 'عمق' in x or 'العمق' in x]
        length2 = length1[0].split(' ')
        length3 = [x for x in length2 if x.isdigit()]
        length = str(length3[0])
        print('length',length)
    except:
        print('Pas de height')
        length = ''
    
    if size == '':
        for size_find in combination_product_size_list1:
            if size_find.find('Product Size') == 0:
                pattern_H = 'H[1-9]+'
                pattern_D = 'D[1-9]+'
                pattern_W = 'W[1-9]+'
                width = [re.findall(pattern_W, com)[0].replace('W', '') for com in combination_product_size_list1 if re.findall(pattern_D, com) != []][0]
                height = [re.findall(pattern_H, com)[0].replace('H', '') for com in combination_product_size_list1 if re.findall(pattern_D, com) != []][0]
                length = [re.findall(pattern_D, com)[0].replace('D', '') for com in combination_product_size_list1 if re.findall(pattern_D, com) != []][0]

    
    
    if width == '':
        width = '__EMPTY__VALUE__'
    if height == '':
        height = '__EMPTY__VALUE__'
    if length == '':
        length = '__EMPTY__VALUE__'
# STYLE  *******************************************   
    style = return_details('النوع :', combination_product_size_list1).replace("\n", " ").replace('♦', '')

    if style == '':
        try:


            for styl in combination_product_size_list1:
                

                if styl.find('Concept') == 0:
                    style = styl.replace('Concept :\n♦', '').replace(':', '').replace('\n', '')


        except:
            style = ''
    if style.find('نوع المنتج') == 0:
        style = style.replace('نوع المنتج', '').replace(':', '')
    print('Style:', style)
    
    if style == '':
        style = '__EMPTY__VALUE__'
#     price = driver.find_element_by_id("product-attribute-specs-table").text
#     print(price)
    #**Overview************************************************
    overview = ''
    try:
        overview = driver.find_element_by_class_name('std').text.split('\n\n')
        overview  = [text.replace('نظرة عامة :', '') for text in overview if 'نظرة عامة' in text][0]
    except:
        overview = ''
        print('ya pas ')
    print('Overview:', overview)
    list_input_overview.append(overview)
#*********************************************************

#**حجم المنتج :************************************************
    product_size = ''
    try:
        product_size = driver.find_element_by_class_name('std').text.split('\n\n')
        product_size  = [text.replace('حجم المنتج :', '') for text in product_size if 'حجم المنتج' in text][0].split('\n')
        
    except:
        product_size = ''
        print('ya pas ')
        
    for i in product_size:
        if i == '':
            product_size.remove(i)
    print('Product_size:', product_size)
#*********************************************************
    list_concatenation.append(f'{sku} {product_name}')

# IMAGES ***************************************************
    list_images = []
    
    
    images = driver.find_elements_by_class_name("gallery-image")# 4 images.
    for i in images:
        list_images.append(i.get_attribute('src'))
    list_input_title.append(product_name)
    list_input_price.append(price)
    list_input_price_was.append(old_price)
    list_input_model.append(detail)
    list_input_color.append(color)
    list_input_size.append(size)
    list_input_in_stock.append(in_stock)
    list_input_sku.append(sku)
    list_input_style.append(style)
    list_input_mater.append(mater)
    list_input_link_url.append(link_url)
    list_input_category1.append(category1)
    list_input_category2.append(category2)
    list_input_category3.append(category3)
    list_input_width.append(width)
    list_input_height.append(height)
    list_input_length.append(length)
    list_input_note.append(note)
    list_tax_class_name.append('Taxable Goods')
    list_visibility.append('Catalog, Search')
    
   
    
    try:
        t = ''
        
        for i, img in enumerate(product_size):
            if i == 0:
                t += str(img.replace('♦', ''))
            
            else:
                t += ', ' + str(img.replace('♦', ''))
        list_input_product_size.append(t)
    except:
        list_input_images2.append('')
    
    try:
        list_input_images1.append(list_images[0])
    except:
        list_input_images1.append('')
        
    try:
        t = ''
        
        for i, img in enumerate(list_images[2:]):
            if i == 0:
                t += str(img)
            if i == 1:
                pass
            else:
                t += ', ' + str(img)
        list_input_images2.append(t)
    except:
        list_input_images2.append('')
        
        
    try:
        list_input_images3.append(list_images[1])
    except:
        list_input_images3.append('')
#     try:
#         list_input_images4.append(list_images[3])
#     except:
#         list_input_images4.append('')
    #df['Images'] = list_images

df = pd.DataFrame({'sku': list_input_sku,
                    'is_in_stock': list_input_in_stock,
                    'name': list_input_title,
                    'estimated_delivery_text': list_input_note,
                   'url_key': list_concatenation,
                   'special_price': list_input_price, 
                   "price":list_input_price_was,
                   'visibility': list_visibility,
                   'tax_class_name': list_tax_class_name,
                   'set_include': list_input_model, 
                   'short_description': list_input_overview, 
                   'set_include': list_input_model,
                   'free_colors': list_input_color,
                   'products_size': list_input_product_size,
                   'ts_dimensions_width': list_input_width,
                   'ts_dimensions_height': list_input_height,
                   'ts_dimensions_length': list_input_length,
                   'style': list_input_style,
                   'raw_materials': list_input_mater,
                   'Images Style': list_input_images3,
                   'Images1': list_input_images1,
                   'Images others': list_input_images2,
#                    'Images4': list_input_images4,
                   'Link url': list_input_link_url,
                   'categories1': list_input_category1,
                   'categories2': list_input_category2,
                   'categories3': list_input_category3,

                  })
#     new_df = pd.concat([df])
df.to_excel( 'midas_update_photo.xlsx', index = True)
```
